```python
import os
import pandas as pd
```

```python
zoc_df = pd.read_csv("/kaggle/input/zocdoc/zocdoc_finalised.csv")
dac_df = pd.read_csv("/kaggle/input/zocdoc/dac_finalised.csv", low_memory=False)
```

```python
zoc_df["facility_address"] = zoc_df["facility_address"].fillna(zoc_df["street_address"])

zoc_df["cred"] = zoc_df["name"].apply(lambda x: x.split(",")[1].strip() 
                                      if len(x.split(",")) > 1 else None)

zoc_df["name"] = zoc_df["name"].apply(lambda x: x.split(",")[0].strip())

zoc_df["postal_code"] = zoc_df["postal_code"].astype(str)

zoc_df = zoc_df[['name', 'cred', 'job_title', 'facility_name', 'facility_address', 'address_locality', 
                 'region', 'postal_code', 'is_dr', 'is_nurse', 'is_assistant']]

zoc_df["is_zoc"] = 1
```

```python
dac_df = dac_df.rename(columns={'street_address': 'facility_address'})

dac_df["postal_code"] = dac_df["postal_code"].apply(lambda x: str(x)[:5])

dac_df = dac_df[['name', 'cred', 'job_title', 'facility_name', 'facility_address', 'address_locality', 
                 'region', 'postal_code', 'is_dr', 'is_nurse', 'is_assistant']]

dac_df["is_zoc"] = 0
```

```python
print(zoc_df.shape)

zoc_df.head()
```

```python
print(dac_df.shape)

dac_df.head()
```

```python
merged = pd.concat([zoc_df, dac_df])

print(merged.shape)

merged.head()
```

```python
zoc_df.to_csv("zoc_final.csv", index=False)
dac_df.to_csv("dac_final.csv", index=False)
merged.to_csv("merged.csv", index=False)
```
